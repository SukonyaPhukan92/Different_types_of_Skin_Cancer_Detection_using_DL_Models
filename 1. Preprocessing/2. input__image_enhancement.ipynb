```python
from PIL import Image,ImageEnhance  
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
def main():
  inPath ="/content/drive/MyDrive/Seventh Semester Project/Skin Cancer Classification/Augmented_Images/actinic keratosis"
  outPath ="/content/drive/MyDrive/Seventh Semester Project/Skin Cancer Classification/Enhanced Images Again/Actinic Keratosis"
  for imagePath in os.listdir(inPath):
    inputPath=os.path.join(inPath, imagePath)
    img=Image.open(inputPath)
    im_out = ImageEnhance.Contrast(img).enhance(1.5)
    fullOutPath=os.path.join(outPath,'Enhanced_'+imagePath)
    im_out.save(fullOutPath)
    print(fullOutPath)
if __name__=='__main__':
  main()
```
